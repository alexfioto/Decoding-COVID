```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import us_states
import fancy_heatmap
```

```python
state_dict = dict(map(reversed, us_states.abbrev_us_state.items()))
```

```python
data_covid_raw = pd.read_csv('../data/covid_metrics_full_time.csv', index_col=0)
```

```python
data_alex_raw = pd.read_csv('../data/sme.csv', index_col=0)
```

```python
data_covid = data_covid_raw.copy()
```

```python
data_mask_political = data_alex_raw.copy()
```

```python
columns_lower = [i.lower() for i in data_mask_political.columns]
data_mask_political.columns = columns_lower
```

```python
data_covid['state'] = data_covid['state'].map(state_dict)
```

```python
data_mask_political.head()
```

```python
data_covid.head()
```

```python
data_covid_mask_political = pd.merge(data_covid,data_mask_political, on='state')
```

```python
data_covid_mask_political.head().T
```

```python
data_covid_mask_political['percent pop affected'] = (data_covid_mask_political['cases'] + data_covid_mask_political['deaths']) / data_covid_mask_political['pop']
```

```python
data_covid_mask_political.to_csv('covid_mask_political_combined.csv')
```
