```python
states = pd.read_csv('./us-states.csv')
```

```python
states['date'] = pd.to_datetime(states['date'])
```

### Slice DataFrame to include 2020-07-01 to 2020-09-01 only

```python
snapshot = states[(states['date']>='2020-07-01') & (states['date']<='2020-09-01')]  
```

```python
snapshot['death_per_day'] = snapshot['deaths'].diff()
snapshot.drop(columns='death_per_day',inplace=True)
```

```python
snapshot
```

```python
new_snapshot = pd.DataFrame()

for state in snapshot['state'].unique():
    one_state_df = snapshot[snapshot['state']==state]
    
    one_state_df['daily_new_cases'] = one_state_df['cases'].diff()
    one_state_df['daily_new_deaths'] = one_state_df['deaths'].diff()
    
    one_state_df['zeroed_cases'] = one_state_df['cases'] - list(one_state_df['cases'])[0]
    one_state_df['zeroed_deaths'] = one_state_df['deaths'] - list(one_state_df['deaths'])[0]
    
    new_snapshot = new_snapshot.append(one_state_df)
```

```python
new_snapshot.fillna(0,inplace=True)
new_snapshot.reset_index(inplace=True)
```

```python
new_snapshot.drop(columns='index',inplace=True)
```
